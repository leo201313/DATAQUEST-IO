```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
star_wars = pd.read_csv('star_wars.csv',encoding='ISO-8859-1')
star_wars.head(10)

%matplotlib inline
```

```python
star_wars.columns
```

```python
star_wars = star_wars[star_wars['RespondentID'].notnull()]
```

```python
star_wars.head()
```

```python
yes_no = {'Yes':True,'No':False}
star_wars['Have you seen any of the 6 films in the Star Wars franchise?'] = star_wars['Have you seen any of the 6 films in the Star Wars franchise?'].map(yes_no)
star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'] = star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'].map(yes_no)
star_wars.head()
```

```python
seen_or_not = {'Star Wars: Episode I  The Phantom Menace':True,
               'Star Wars: Episode II  Attack of the Clones':True,
               'Star Wars: Episode III  Revenge of the Sith':True,
               'Star Wars: Episode IV  A New Hope':True,
               'Star Wars: Episode V The Empire Strikes Back':True,
               'Star Wars: Episode VI Return of the Jedi':True,
               np.nan:False
              }
# np.nan and np.NaN is the same

name_change ={"Which of the following Star Wars films have you seen? Please select all that apply.": "seen_1",
              "Unnamed: 4": "seen_2",
              "Unnamed: 5": "seen_3",
              "Unnamed: 6": "seen_4",
              "Unnamed: 7": "seen_5",
              "Unnamed: 8": "seen_6"}

star_wars.rename(columns=name_change,inplace=True)
# rename the columns

for cols in star_wars.columns[3:9]:
    star_wars[cols] = star_wars[cols].map(seen_or_not)
    
star_wars.head(10)
```

```python
star_wars.iloc[:,9:15] = star_wars.iloc[:,9:15].astype(float)
name_trans = {'Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.':'ranking_1',
              'Unnamed: 10':'ranking_2',
              'Unnamed: 11':'ranking_3',
              'Unnamed: 12':'ranking_4',
              'Unnamed: 13':'ranking_5',
              'Unnamed: 14':'ranking_6'
             }
star_wars.rename(name_trans,axis=1,inplace=True)
star_wars.iloc[:,9:15].head(10)
```

```python
rank_mean = star_wars.iloc[:,9:15].mean(axis=0)
rank_mean
```

```python
rank_mean.plot.bar()
```

It seems the third movie gets the most high point which says it is the most unfavorited one while the fifth movie gets the lowest point which says it is the most favorite one. Maybe with the movie technology is developing, the movie is seemed more and more real which is important for a fection movie series. And maybe the fifth one is edited well which cause the sixth one also has a low point, though it is not well.

```python
star_wars['total_seen'] = star_wars.iloc[:,3:9].sum(axis=1)
star_wars['total_seen'] 
```

```python
seens_grouped = star_wars.pivot_table(values=star_wars.columns[9:15],columns='total_seen')
seens_grouped
```

```python
fig = plt.figure()
for ele in range(7):
    ax = fig.add_subplot(2,4,ele+1)
    ax = seens_grouped[ele].plot.bar(ylim=(0,6))
    col = str(ele)
    ax.set_title('total seen: '+col)
    ax.tick_params(right='off',left='off',top='off',bottom='off')
    if ele in range(3):
        ax.tick_params(labelbottom='off')
    if ele not in [0,4]:
        ax.tick_params(labelleft='off')
plt.show()
```

For people who have seen at least two of this series of movies are more likely to prefer the fifth one of this series.For people who only see no movie, it is ridiculous to give a point for any one of these movies, and they may just point the movie by the informations they got from others or Internet.And for the people who only see one movie also take no efford for them to point the movie cause they don't have another one to compare with.  

```python
fan_group = star_wars[star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?']==True].mean()
unfan_group = star_wars[star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?']==False].mean()
fig1 = plt.figure()
ax1 = fig1.add_subplot(1,2,1)
ax1 = fan_group[9:15].plot.bar()
ax1.set_title('Fans')
ax2 = fig1.add_subplot(1,2,2)
ax2 = unfan_group[9:15].plot.bar()
ax2.set_title('UnFans')
plt.show()
```

For both fans and unfans, the most favorite one is the same -- the fifth one.However we can also find that the data difference in fans group are more huge(which means the fans point higher point for unfavorite movies and point lower point for favorite ones)
